SCATTER PLOT

```python
ms.scatterplot(x="alcohol",y="total",data=d,color='blue')
```

LINE PLOT

INFERENCE: As the rate of alcohol consumption increasing, the total accedents are also increasing.

```python
ms.lineplot(x="ins_premium",y="ins_losses",data=d,color='red')
```

INFERENCE: As the rate of insurence_premium increasing, the insurrence_losses are also increasing.

DISPLOT

```python
ms.displot(d['speeding'])
```

INFERENCE: Maximum car crashes are happened in between 4-5 speed 

DISTRIBUTION PLOT

```python
ms.distplot(d['speeding'])
```

```python
d['speeding'].mean()
```

```python
d['speeding'].median()
```

```python
d['speeding'].mode()
```

INFERENCE:The distribution is almost symmetrical

RELPLOT

```python
ms.relplot(x="alcohol",y="total",data=d,hue="abbrev")
```

INFERENCE: The plotting is done based on states 

BARPLOT

```python
import matplotlib.pyplot as p
```

```python
p.subplots(figsize=(106,20))
ms.barplot(x="abbrev",y="total",data=d)
```

COUNTPLOT

```python
p.subplots(figsize=(16,10))
ms.countplot(y="not_distracted",data=d,orient='h')
```

INFERENCE: The Not_distracted cases are more at 14.094

```python
d['not_distracted'].mode()
```

JOINPLOT

```python
ms.jointplot(x="ins_premium",y="ins_losses",data=d)
```

INFERENCE: As the Insurance Premium is taken but the Insurance losses are also high in state wise

HEATMAP FOR CORRELATION

```python
ms.heatmap(d.corr(),annot=True,cmap="YlGnBu")
```

INFERENCE: Total is highly correlated with Alcohol to test car crash

PAIRPLOT

```python
ms.pairplot(d)
```

INFERENCE: Above methods tells us the Trend between all Numerical Analasys

BOXPLOT

```python
p.subplots(figsize=(16,10))
ms.boxplot(x="total",y="abbrev",data=d)
```

INFERENCE: As the states of USA are unique so as mean, median and mode are equal. It indicates only one value
